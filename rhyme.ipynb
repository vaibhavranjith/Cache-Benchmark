```
import nltk
from nltk.corpus import treebank
import cmudict
# nltk.download('punkt')
# nltk.download('treebank')
nltk.download('cmudict')
```

```
def get_rhymes(the_word):
  poem=[ele[0] for ele in cmudict.entries()]
  arpabet=nltk.corpus.cmudict.dict()
  last_phoneme_of_given_word=arpabet[the_word][0][-1]
  second_last_last_phoneme_of_given_word=arpabet[the_word][0][-2]
  rhymes=[]
  for ele in poem:
    try:
      phonemes=arpabet[ele][0]
      sz=len(phonemes)
      last_phoneme=phonemes[-1];
      second_last_phoneme=phonemes[-2]
      if(last_phoneme==last_phoneme_of_given_word 
         and second_last_last_phoneme_of_given_word==second_last_phoneme ):
        rhymes.append(ele)
    except:
      pass
  if(len(rhymes)==0):
    return "Sorry have'nt heard this word before dawg!"
  print(rhymes)
  return rhymes
```

```

grammar1 = nltk.CFG.fromstring("""
 S -> NP VP
PP -> P NP
NP -> Det N | Det N PP | 'I'
VP -> V NP | VP PP
Det -> 'an' | 'my'
N -> 'elephant' | 'pajamas'
V -> 'shot'
P -> 'in'
  """)
```

```
sent="I shot an elephant in my pajamas ".split()
rd_parser=nltk.RecursiveDescentParser(grammar1)
for tree in rd_parser.parse(sent):
  print(tree) 
```
